```python
class Money_In:
    def Income(self):
        rental = int(input('How much do you charge for rent? '))
        laundry = int(input('laundry cost? '))
        storage = int(input('storage cost? '))
        misc = int(input('Misc cost? '))

        total_income = rental + laundry + storage + misc
        print(f'Your total income is {total_income}')
        return total_income


class Money_Out:
    def Expense(self):
        tax = int(input('Enter Tax amount: '))
        insurance = int(input('Monthly Insurance amount: '))
        utilities = int(input('Enter utilities: '))
        hoa = int(input('Enter HOA fees: '))
        maintenance = int(input('Enter maintenance expenses: '))
        vacancy = int(input('Enter Vacancy amt: '))
        capex = int(input('Enter total cost to replace items: '))
        property_management = int(input('Enter property management Expense: '))
        mortgage = int(input('Mortgage: '))

        total_expense = tax + insurance + utilities + hoa + maintenance + vacancy + capex + property_management + mortgage
        print(f'Your total expenses are {total_expense}')
        return total_expense


class Cash_flow(Money_In, Money_Out):
    def C_F(self):
        w = Money_In()
        x = Money_Out()
        y = w.Income() - x.Expense()
        z = y * 12
        print(f'Your monthly cash flow is {y}. ')
        print(f'your yearly cash flow is {z}. ')

        down_payment = int(input('Down payment? '))
        closing = int(input('Closing cost? '))
        rehab = int(input('Cost to rehab? '))
        misc = int(input('Misc investment cost? '))

        invest = down_payment + closing + rehab + misc
        print(f'You have spent a total of {invest} dollars on investments')
        roi = (z/invest) * 100
        print(f'Your return on investment is {roi}%')


def run():
    while True:
        begin = input('Do you wish to start the program? Y = yes, N = no ')
        if begin.lower() == 'y' or begin.lower() == 'yes':
            cash = Cash_flow()
            cash.C_F()
        elif begin.lower() == 'n' or begin.lower() == 'no':
            print('Have a good day')
        else:
            print('Please enter a valid option')

        print('Thank you for using our calculator')
        break


run()
```
